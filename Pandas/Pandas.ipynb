# **Series**

```python
import pandas as pd
x=[3,4,5,6]
var = pd.Series(x,index=['a','s','d','f'],dtype="float",name="python")
print(var)
print(type(var))
print(var['d'])
```

```python
dic = {"name":['python','c','c++','java'], "por":[12,13,14,15], "rank":[1,4,3,2]}
var1 = pd.Series(dic)
print(var1)
```

```python
s = pd.Series(12, index=[1,2,3,4,5,6,7])
print(s)
print(type(s))
```

```python
s1 = pd.Series(12, index=[1,2,3,4,5,6,7])
s2 = pd.Series(12, index=[1,2,3,4])

print(s1+s2)
```

# **Data Frame**

```python
l=[1,2,3,4,5,6]
var11 = pd.DataFrame(l)
print(type(var11))
print(var11)
```

```python
d = {"a":[1,2,3,14,5], "s":[1,2,3,4,5], "x":[1,2,3,4,5], 1:[1,2,3,4,5]}

var12 = pd.DataFrame(d)
print(type(var12))
print(var12)
print(var12["a"][3])
```

```python
list_1 = [[1,2,3,4,5],[11,12,13,14,15]]
var13 = pd.DataFrame(list_1)
print(type(var13))
print(var13)
```

```python
sr = {"s":pd.Series([1,2,3,4]), "r":pd.Series([1,2,3,4])}
var14 = pd.DataFrame(sr)
print(type(var14))
print(var14)
```

## **Arithmetic Operations**

```python
var31 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var31
```

```python
var31["C"] = var31["A"]+var31["B"]
var31
```

```python
var32 = pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
var32
```

```python
var32["python"] = var32["A"] <= 20
var32
```
